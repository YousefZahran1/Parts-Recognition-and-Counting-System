```python
"""#Importing Libraries"""

import os
import glob
import cv2
import numpy as np
from PIL import Image
from sklearn import svm
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV
import math
import h5py

"""#Preprocessing Functions"""

# Preprocessing function
def preprocess_images(folders, folder_path, image_size):
    image_names_list = []
    for folder in folders:
        for f in glob.glob(os.path.join(folder_path, folder, '*.jpg')):
            image_names_list.append(f)

    for image in image_names_list:
        pre_image = Image.open(image).convert('L')
        new_image = pre_image.resize(image_size)
        new_image.save(image)
        
# Preprocessing for test
test_folders = ['Part_A', 'Part_B', 'Part_C', 'Part_D']
preprocess_images(test_folders, 'C:/Users/youse/Downloads/TPR project/TPR project/Test', (150, 150))

# Preprocessing for training
train_folders = ['Part_A', 'Part_B', 'Part_C', 'Part_D']
preprocess_images(train_folders, 'C:/Users/youse/Downloads/TPR project/TPR project/Train', (150, 150))

"""#Data Loading Functions"""

# Load data function
def load_data(folders, folder_path, image_size):
    data = []
    for folder in folders:
        path = os.path.join(folder_path, folder)
        class_num = folders.index(folder)
        for image in os.listdir(path):
            image_array = cv2.imread(os.path.join(path, image), cv2.IMREAD_GRAYSCALE)
            new_array = cv2.resize(image_array, image_size)
            data.append([new_array, class_num])

    return data

# Extract features and labels function
def extract_features_labels(data, image_size):
    X = []
    Y = []
    for features, label in data:
        X.append(features)
        Y.append(label)

    X = np.array(X)
    Y = np.array(Y)

    # Reshape the numpy array
    X = X.reshape(X.shape[0], -1)

    # Normalize to 0-1
    X = X / 255.0

    return X, Y

# Split train and test data function
def split_train_test_data(X, Y, test_size):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=42)
    return X_train, Y_train, X_test, Y_test

"""#SVM Model Training and Evaluation"""

# Train the SVM model function
def train_svm(X_train, Y_train):
    # Define the parameter grid for grid search
    param_grid = {
      'C': [1, 10, 100, 1000],  # Penalty parameter C
      'gamma': [0.001, 0.01, 0.1, 1],  # Kernel coefficient for 'rbf'
      'kernel': ['rbf', 'sigmoid', 'linear']  # Kernel options
      }

    # Perform grid search to find the best hyperparameters
    grid_search = GridSearchCV(svm.SVC(class_weight='balanced'), param_grid, cv=5)
    grid_search.fit(X_train, Y_train)

    # Get the best SVM classifier
    best_svm_classifier = grid_search.best_estimator_

    # Train the best SVM classifier
    best_svm_classifier.fit(X_train, Y_train)

    return best_svm_classifier

# Evaluate the model function
def evaluate_model(classifier, X_test, Y_test):
    Y_pred = classifier.predict(X_test)

    accuracy = accuracy_score(Y_test, Y_pred)
    classification_report_output = classification_report(Y_test, Y_pred)
    confusion_matrix_output = confusion_matrix(Y_test, Y_pred)

    return accuracy, classification_report_output, confusion_matrix_output

"""#Visualization Functions"""

# Plot PCA explained variance function
def plot_pca(pca, X_train):
    plt.figure(figsize=(8, 6))
    plt.plot(np.cumsum(pca.explained_variance_ratio_))
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.title('PCA - Explained Variance')
    plt.grid(True)
    plt.show()

def show_eigenfaces(pca, image_size):
    eigenfaces = pca.components_.reshape(-1, image_size[0], image_size[1])
    num_components = len(eigenfaces)
    
    # Reshape eigenfaces to match the desired format
    eigenfaces = eigenfaces.reshape((num_components, image_size[0], image_size[1]))
    
    # Call your image_grid function passing the eigenfaces
    image_grid(eigenfaces, image_size[0], image_size[1], cols=5)

# Display image grid
def image_grid(D, H, W, cols=10, scale=1):
    n = np.shape(D)[0]
    rows = int(math.ceil((n + 0.0) / cols))
    fig = plt.figure(1, figsize=[scale * 20.0 / H * W, scale * 20.0 / cols * rows], dpi=300)
    for i in range(n):
        plt.subplot(rows, cols, i + 1)
        fig = plt.imshow(np.reshape(D[i, :], [H, W]), cmap=plt.get_cmap("gray"))
        plt.axis('off')

# Display image grid
def show_images(data, folders):
    num_classes = len(folders)
    plt.figure(figsize=(10, 10))
    for i in range(num_classes):
        class_data = [x[0] for x in data if x[1] == i]
        class_name = folders[i]
        for j in range(5):
            plt.subplot(num_classes, 5, i * 5 + j + 1)
            plt.imshow(class_data[j], cmap='gray')
            plt.title(class_name)
            plt.axis('off')
    plt.tight_layout()
    plt.show()

# Plot confusion matrix function
def plot_confusion_matrix(confusion_matrix_output, classes):
    plt.figure(figsize=(8, 6))
    sns.heatmap(confusion_matrix_output, annot=True, cmap='Blues', fmt='g', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title('Confusion Matrix')
    plt.show()

"""#Feature Extraction and Analysis Function"""

# Feature extraction using PCA
def feature_extraction(pca_model, train, test):
    X_pca_train = pca_model.transform(train)
    X_pca_test = pca_model.transform(test)
    return X_pca_train, X_pca_test

# Find the mean image
def calculate_mean_image(data):
    num_images = len(data)
    mean_image = np.zeros(data[0][0].shape)
    for features, _ in data:
        mean_image += features
    mean_image /= num_images
    return mean_image

"""#Data Loading and Preprocessing"""

# Load data
image_size = (150, 150)
train_data = load_data(train_folders, 'C:/Users/youse/Downloads/TPR project/TPR project/Train', image_size)
test_data = load_data(test_folders, 'C:/Users/youse/Downloads/TPR project/TPR project/Test', image_size)

# Extract features and labels
X_train, Y_train = extract_features_labels(train_data, image_size)
X_test, Y_test = extract_features_labels(test_data, image_size)

assert len(X_train) == len(Y_train), "Mismatch in train data and labels"
assert len(X_test) == len(Y_test), "Mismatch in test data and labels"

# Show training images
show_images(train_data, train_folders)

# Show testing images
show_images(test_data, test_folders)

"""#Data Storage and Loading Using H5"""

# Store test data to H5 file
if not os.path.exists('data_test_svm.h5'):
    hf_test = h5py.File('data_test_svm.h5', 'w')
    hf_test.create_dataset('X_test', data=X_test)

# Store train data to H5 file
if not os.path.exists('data_train_svm.h5'):
    hf_train = h5py.File('data_train_svm.h5', 'w')
    hf_train.create_dataset('X_train', data=X_train)

# Load H5 files
h5_train = h5py.File('data_train_svm.h5', 'r')
h5_test = h5py.File('data_test_svm.h5', 'r')

# Load the datasets from H5 files
X_train = np.array(h5_train["X_train"])
X_test = np.array(h5_test["X_test"])

"""#PCA Dimensionality Reduction and SVM Model Training"""

print("X_train shape:", np.array(X_train).shape)
print("Y_train shape:", np.array(Y_train).shape)
print("X_test shape:", np.array(X_test).shape)
print("Y_test shape:", np.array(Y_test).shape)

# Split train and test data
X_train, Y_train, X_test, Y_test = split_train_test_data(X_train, Y_train, test_size=0.2)

pca = PCA(n_components=40)
pca.fit(X_train)

# Perform dimensionality reduction with PCA
X_pca_train, X_pca_test = feature_extraction(pca, X_train, X_test)

# Find the mean image
mean_image = calculate_mean_image(train_data)
plt.imshow(mean_image, cmap='gray')
plt.title('Mean Image')
plt.axis('off')
plt.show()

# Showing the eigenfaces
show_eigenfaces(pca, image_size)
plt.show()

# Train the SVM model
best_svm_classifier = train_svm(X_pca_train, Y_train)

"""#Model Evaluation and Visualization"""

# Predict labels for test data using the best SVM classifier
Y_pred_best = best_svm_classifier.predict(X_pca_test)

# Evaluate the best SVM classifier
accuracy_best, classification_report_best, confusion_matrix_best = evaluate_model(best_svm_classifier, X_pca_test, Y_test)
# Print the results for the best SVM classifier
print("Best SVM Accuracy:", accuracy_best)
print("Best SVM Classification Report:")
print(classification_report_best)
print("Best SVM Confusion Matrix:")
print(confusion_matrix_best)

# Plot PCA explained variance
plot_pca(pca,X_pca_train)

# Calculate intra-class distance
intra_distances = np.sum(np.power(X_pca_train[1, :] - X_pca_test[1, :], 2))

# Calculate inter-class distance
inter_distances = np.sum(np.power(X_pca_train[1, :] - X_pca_test[6, :], 2))

# Print the distances without decimal places
print("Intra-class distance: %d" % intra_distances)
print("Inter-class distance: %d" % inter_distances)

# Plot confusion matrix
class_names = train_folders
plot_confusion_matrix(confusion_matrix_best, class_names)

print("Shape of training set: {}".format(X_train.shape))
print("Shape of training labels: {}".format(Y_train.shape))
print("Shape of testing set: {}".format(X_test.shape))
print("Shape of transformed training set (pca): {}".format(X_pca_train.shape))
print("Shape of transformed testing set (pca): {}".format(X_pca_test.shape))
```
